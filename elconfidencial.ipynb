## Scrapping de "El Confidencial"

```python
import time
import feedparser
from bs4 import BeautifulSoup
```

```python
url_nac = "https://rss.elconfidencial.com/espana/"
url_inter="https://rss.elconfidencial.com/mundo/"
url_cult="https://rss.elconfidencial.com/cultura/"

urls=[url_nac, url_inter, url_cult]

categoria=["Nacional", "Internacional", "Cultura"]
```

```python
noticias=[]

for url, cat in zip(urls, categoria):
    feed = feedparser.parse(url)

    for entry in feed.entries:

        contenido_html=entry.content[0].value

        contenido=BeautifulSoup(contenido_html, "html.parser")

        noticia={}

        noticia["Link"]=entry.link
        noticia["Periódico"]= "El Confidencial"
        noticia["Fecha"]=time.strftime("%Y-%m-%d", entry.published_parsed)
        noticia["Título"]=entry.title
        noticia["Subtítulo"]=entry.summary
        noticia["Categoría"]=cat
        noticia["Contenido"]=contenido.text
        
        noticias.append(noticia)
```

```python
import json
import os

ruta_archivo = "data/elconfidencial.json"

if os.path.exists(ruta_archivo):
    with open(ruta_archivo, 'r', encoding='utf-8') as archivo:
        noticias_anteriores = json.load(archivo)
    
    links_anteriores = [noticia['Link'] for noticia in noticias_anteriores]

    for noticia in noticias:
        if noticia['Link'] not in links_anteriores:
            noticias_anteriores.append(noticia)

    with open(ruta_archivo, 'w', encoding = 'utf-8') as archivo:
        json.dump(noticias_anteriores, archivo, indent = 1, ensure_ascii = False)
    
else:
    with open(ruta_archivo, 'w', encoding = 'utf-8') as archivo:
        json.dump(noticias, archivo, indent = 1, ensure_ascii = False)
```
