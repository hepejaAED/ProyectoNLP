## Notebook Scrappeo 20minutos (https://www.20minutos.es/rss)

```python
import feedparser

# URL del feed RSS de 20minutos
url_inter = 'https://www.20minutos.es/rss/internacional/'
url_nac = 'https://www.20minutos.es/rss/nacional/'
url_cult = 'https://www.20minutos.es/rss/cultura/'
```

```python
import time
from bs4 import BeautifulSoup

urls = [url_inter, url_nac, url_cult]
categorias = ["Internacional", "Nacional", "Cultura"]

noticias = []

for url, cat in zip(urls, categorias):

    feed = feedparser.parse(url)

    for entry in feed.entries:
        if "content" in entry.keys():
            noticia = {
            "Link": entry.link,
            "Periódico": "20minutos",
            "Fecha": time.strftime("%Y-%m-%d", entry.published_parsed),
            "Título": entry.title,
            "Subtítulo": entry.summary,
            "Categoría": cat,
            "Contenido": BeautifulSoup(entry.content[0].value, 'html.parser').text
            }

            noticias.append(noticia)
```

Guardamos las noticias extraídas en un archivo `.json` con el nombre `20minutos.json`.

```python
import json
import os

ruta_archivo = "data/20minutos.json"

if os.path.exists(ruta_archivo):
    with open(ruta_archivo, 'r', encoding='utf-8') as archivo:
        noticias_anteriores = json.load(archivo)
    
    links_anteriores = [noticia['Link'] for noticia in noticias_anteriores]

    for noticia in noticias:
        if noticia['Link'] not in links_anteriores:
            noticias_anteriores.append(noticia)

    with open(ruta_archivo, 'w', encoding = 'utf-8') as archivo:
        json.dump(noticias_anteriores, archivo, indent = 1, ensure_ascii = False)
    
else:
    with open(ruta_archivo, 'w', encoding = 'utf-8') as archivo:
        json.dump(noticias, archivo, indent = 1, ensure_ascii = False)
```
